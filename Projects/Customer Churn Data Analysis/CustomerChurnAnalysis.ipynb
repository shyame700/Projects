```python
import seaborn as sns
import matplotlib.pyplot as plt
#Churn Distribution
sns.countplot(data=df, x='Churn')
plt.title("Churned vs Retained customers")
plt.xticks([0,1],['Not Churned','Churned'])
plt.ylabel("Number of Customers")
plt.show()
#Conclusion
churn_rate=df['Churn'].mean()
print(f"Overall Churn Rate: {churn_rate:.2%}")
```

```python
# Categorical Feature vs Churn
#Gender vs Churn
sns.countplot(data=df, x='Gender', hue='Churn')
plt.title("Churn by Gender")
plt.show()
```

```python
#Subscription Type vs churn
sns.countplot(data=df, x='Subscription Type', hue='Churn')
plt.title("Churn by Subscription Type")
plt.show()
```

```python
#Contract Length vs Churn
sns.countplot(data=df, x='Contract Length', hue='Churn')
plt.title("Churn by Contract Length")
plt.show()
```

```python
num_cols=['Age','Tenure','Usage Frequency','Support Calls','Payment Delay','Total Spend', 'Last Interaction']
for col in num_cols:
    sns.boxplot(data=df,x='Churn', y=col)
    plt.title(f'{col} by Churn')
    plt.show()
```

```python
import numpy as np
corr_df=df.drop(columns=['CustomerID', 'Gender', 'Subscription Type', 'Contract Length'])
corr =corr_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
summary = df.groupby('Subscription Type')['Churn'].agg(['count', 'mean']).rename(columns={'mean': 'Churn Rate'})
summary['Churn Rate'] = summary['Churn Rate'].apply(lambda x: f"{x:.2%}")
summary
```
