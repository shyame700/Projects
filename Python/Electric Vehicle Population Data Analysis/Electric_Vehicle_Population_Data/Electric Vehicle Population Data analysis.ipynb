```python
df['Model'] = df['Model'].fillna(df['Model'].mode()[0])
df['Vehicle Location'] = df['Vehicle Location'].fillna('Unknown')
```

```python
df['Legislative District'] = df['Legislative District'].fillna(0)
df['Electric Utility'] = df['Electric Utility'].fillna('Unknown')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20, 10))

# Bar plot
plt.subplot(1, 2, 1)
plt.title("Electric Vehicle Type Distribution")
sns.countplot(x='Electric Vehicle Type', data=df)
plt.xticks(rotation=45)

# Pie chart
plt.subplot(1, 2, 2)
plt.title("Electric Vehicle Type Distribution (%)")
ev_counts = df['Electric Vehicle Type'].value_counts()
plt.pie(ev_counts, labels=ev_counts.index, autopct="%0.0f%%")

plt.tight_layout()
plt.show()
```

```python
dist_year=df.groupby(['Model Year','Electric Vehicle Type'])['VIN (1-10)'].count().unstack()
dist_year
```

```python
dist_year=df.groupby(['Model Year','Electric Vehicle Type'])['VIN (1-10)'].count().unstack()
dist_year.plot(kind="bar",stacked=True,figsize=(20,10))
plt.title("Electric Vehicle Type Over the Years");
plt.xlabel("Years")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(20,10))
ev=df['County'].value_counts().reset_index()
ev.columns=['Country','Count']
ev=ev.nlargest(10,"Count")
plt.title("Top 10 EV Countries")
plt.xlabel('Country')
plt.ylabel('Number of Vehicles');
sns.barplot(data=ev, x='Country',y='Count')
```

```python
plt.figure(figsize=(20,10))
ev=df['City'].value_counts().reset_index()
ev.columns=['City','Count']
plt.title("Top 10 EV Cities")
plt.xlabel('City')
plt.ylabel('Number of Vehicles');
ev=ev.nlargest(10,"Count")
sns.barplot(data=ev, x="City",y='Count')
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ev=df['Make'].value_counts().reset_index()
ev.columns=['Make','Count']
plt.title("Top 10 Manufacturers")
plt.xlabel('Make')
plt.ylabel('Number of Vehicles');
ev=ev.nlargest(10,"Count")
sns.barplot(data=ev, x="Make",y='Count');


plt.subplot(1,2,2)
plt.title("Top 10 Manufacturers")
ev=ev.nlargest(10,"Count")
plt.pie(ev.Count, labels=ev.Make.unique(), autopct="%0.0f%%");
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
ev=df['Model'].value_counts().reset_index()
ev.columns=['Model','Count']
plt.title("Top 10 EV Models")
plt.xlabel('Model')
plt.ylabel('Number of Vehicles');
ev=ev.nlargest(10,"Count")
sns.barplot(data=ev, x="Model",y='Count');


plt.subplot(1,2,2)
ev=ev.nlargest(10,"Count")
plt.title("Top 10 EV Models")
plt.pie(ev.Count, labels=ev.Model.unique(), autopct="%0.0f%%");
```
